```python
import pandas as pd 
```

```python
url = "https://github.com/PacktWorkshops/The-Data-Science-Workshop/blob/master/Chapter01/Dataset/dataset_44_spambase.csv"
```

```python
df=pd.read_csv('dataset_44_spambase.csv')
```

```python
df
```

```python
target= df.pop('class')
```

```python
print(target)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
seed = 168
```

```python
rf_model= RandomForestClassifier(random_state=seed)
```

```python
rf_model.fit(df,target)
```

```python
preds=rf_model.predict(df)
```

```python
print(preds)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_score= accuracy_score(target,preds)
```

```python
print(acc_score)
```
